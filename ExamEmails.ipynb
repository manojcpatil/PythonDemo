```python
import smtplib
from os.path import basename
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
import pandas as pd
import numpy as np

Data=pd.read_excel(input("Enter the data file"),sheet_name="STUDDATA")

#The mail addresses and password
sender_address = '******@nmu.ac.in'
sender_pass = '******'
subject_title= 'NSD-2021 Attendance Certificate'
mail_content = '''Thank you for being part of NSD-2021 celebration.
We were thrilled by the attendance we had at such short notice. Thanks to all those of you who participated and helped make it a success.
We’d love to hear your feedback! In case if you have missed 'Online Popular Lectures by Mr. Mule and Mr. Mookherjee' it is available at https://youtu.be/-HDMVTog_ds

We thank you again for registering and participating for NSD Celebration. Please let us know if you have any comments or questions and email us and we’ll get back to you soon.
We look forward to bringing you even more programs/presentations in the future!

For any type of query about the certificates, please, fill free to contact on kirteekamalja@gmail.com .

Regards,
-Organizing Committee
 National Statistics Day-2020 Celebration
 Department of Statistics
 School of Mathematical Sciences
 Kavayitri Bahinabai Chaudhari North Maharashtra University, Jalgaon
'''

session = smtplib.SMTP('smtp.office365.com',587)
session.starttls() #enable security
session.login(sender_address, sender_pass) #login with mail_id and password
# session = smtplib.SMTP('smtp.gmail.com', 587) #use gmail with port

for i in np.arange(0,Data.shape[0]):
    mail_content1="Dear Participant,\n"+str(Data.Name[i])+",\n"+str(Data.College[i])+"\n\n"+mail_content
    
    receiver_address = Data.EmailAddress[i]
    #Setup the MIME
    message = MIMEMultipart()
    message['From'] = sender_address
    message['To'] = receiver_address
    message['Subject'] = subject_title

    #The body and the attachments for the mail
    message.attach(MIMEText(mail_content1, 'plain'))

    #Create SMTP session for sending the mail
    text = message.as_string()
    session.sendmail(sender_address, receiver_address, text)
    print('Mail Sent to',i," ",str(Data.Name[i]))

session.quit()
```
