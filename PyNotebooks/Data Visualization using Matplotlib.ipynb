```python
#!pip install matplotlib
import numpy as np, pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
iris=pd.read_csv("Data/Iris.csv")
iris
```

```python
fig, ax = plt.subplots()
ax.plot(iris.SepalLengthCm,iris.PetalLengthCm,color='red')
plt.show()
```

```python
fig, ax = plt.subplots()
iris.target=np.array([[0]*50,[1]*50,[2]*50])
ax.scatter(iris.SepalLengthCm,iris.PetalLengthCm,marker="*",c=iris.target,label="SepalVsPetal")
ax.scatter(iris.SepalLengthCm,iris.SepalWidthCm,marker="o",c=iris.target,label="SepalLengthVsWidth")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend()
plt.title("Sepal Length Vs Petal Length")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie([15,12],labels=["Female","Male"])
plt.legend()
plt.title("Pie Chart")
plt.show()
```

```python
fig, ax = plt.subplots()
x=np.random.normal(10,3,1000)
ax.hist(x,bins=200,label="Normal Distribution",density=False)
plt.legend()
plt.title("Histogram")
plt.show()
```
