```python
import nbformat

# Define the Markdown and Python code as strings
markdown_content = """
## 3. Even or Odd Checker

Create a Python program that prompts the user to enter an integer and checks if it's an even or odd number. Display a message indicating whether the number is even or odd.
"""
python_code = """
# Python Program for Even or Odd Checker
user_number = int(input("Enter an integer: "))

if user_number % 2 == 0:
    print(f"{user_number} is an even number.")
else:
    print(f"{user_number} is an odd number.")
"""

markdown_content = """

# Create a Jupyter Notebook structure
notebook_content = nbformat.v4.new_notebook()
notebook_content.cells.append(nbformat.v4.new_markdown_cell(source=markdown_content))
notebook_content.cells.append(nbformat.v4.new_code_cell(source=python_code))

# Save the notebook content to a file
with open('even_odd_checker.ipynb', 'w', encoding='utf-8') as notebook_file:
    nbformat.write(notebook_content, notebook_file)

print("Jupyter Notebook 'even_odd_checker.ipynb' created successfully.")
```

```python
def addcode(markdown_content, python_code, notebook_file_name='even_odd_checker.ipynb'):
    # Create a Jupyter Notebook structure
    notebook_content = nbformat.v4.new_notebook()
    notebook_content.cells.append(nbformat.v4.new_markdown_cell(source=markdown_content))
    notebook_content.cells.append(nbformat.v4.new_code_cell(source=python_code))
    
    # Save the notebook content to a file
    with open(notebook_file_name, 'w', encoding='utf-8') as notebook_file:
        nbformat.write(notebook_content, notebook_file)

    print(f"Jupyter Notebook '{notebook_file_name}' created successfully.")
```

```python
markdown_content = """
### 1. List Manipulation

#### a. Even Numbers in a List
"""
python_code = """
def filter_even_numbers(input_list):
    return [num for num in input_list if num % 2 == 0]

# Example usage:
numbers_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
even_numbers = filter_even_numbers(numbers_list)
print(even_numbers)
"""

addcode(markdown_content, python_code, notebook_file_name='Practical_02_Q1.ipynb')
```

```python
markdown_content = """

#### b. Reverse List In-Place
"""

python_code = """
def reverse_list_in_place(input_list):
    input_list.reverse()

# Example usage:
original_list = [1, 2, 3, 4, 5]
reverse_list_in_place(original_list)
print(original_list)
"""
addcode(markdown_content, python_code, notebook_file_name='Practical_02_Q2.ipynb')
markdown_content = """

### 2. Tuple Operations

#### a. Sort Tuple Alphabetically
"""

python_code = """
def sort_tuple_alphabetically(input_tuple):
    sorted_tuple = tuple(sorted(input_tuple))
    print(sorted_tuple)

# Example usage:
string_tuple = ('banana', 'apple', 'orange', 'grape')
sort_tuple_alphabetically(string_tuple)
"""
addcode(markdown_content, python_code, notebook_file_name='Practical_02_Q3.ipynb')
markdown_content = """

#### b. Find Index in Tuple
"""

python_code = """
def find_element_index(input_tuple, element):
    try:
        index = input_tuple.index(element)
        print(f"The index of '{element}' is: {index}")
    except ValueError:
        print(f"'{element}' not found in the tuple.")

# Example usage:
my_tuple = (10, 20, 30, 40, 50)
find_element_index(my_tuple, 30)
"""
addcode(markdown_content, python_code, notebook_file_name='Practical_02_Q4.ipynb')
markdown_content = """

### 3. Dictionary Operations

#### a. Merge Dictionaries with Age Combination
"""

python_code = """
def merge_dictionaries(dict1, dict2):
    result_dict = dict1.copy()
    
    for key, value in dict2.items():
        if key in result_dict:
            result_dict[key] += value
        else:
            result_dict[key] = value
            
    return result_dict

# Example usage:
dict1 = {'Alice': 25, 'Bob': 30}
dict2 = {'Bob': 35, 'Charlie': 40}
merged_dict = merge_dictionaries(dict1, dict2)
print(merged_dict)
"""
addcode(markdown_content, python_code, notebook_file_name='Practical_02_Q5.ipynb')
markdown_content = """
#### b. Check if Key Exists in Dictionary
"""

python_code = """
def check_key_existence(input_dict, key):
    if key in input_dict:
        print(f"'{key}' exists in the dictionary.")
    else:
        print(f"'{key}' does not exist in the dictionary.")

# Example usage:
employee_dict = {'Alice': 25, 'Bob': 30, 'Charlie': 40}
check_key_existence(employee_dict, 'Bob')
"""
addcode(markdown_content, python_code, notebook_file_name='Practical_02_Q6.ipynb')
markdown_content = """

### 4. Set Operations

#### a. Common Elements in Sets
"""

python_code = """
def find_common_elements(set1, set2):
    return set1.intersection(set2)

# Example usage:
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
common_elements = find_common_elements(set_a, set_b)
print(common_elements)
"""
addcode(markdown_content, python_code, notebook_file_name='Practical_02_Q7.ipynb')
markdown_content = """

#### b. Remove Duplicates from List and Convert to Set
"""

python_code = """
def remove_duplicates_and_convert_to_set(input_list):
    unique_set = set(input_list)
    return unique_set

# Example usage:
original_list = [1, 2, 2, 3, 4, 4, 5]
unique_set = remove_duplicates_and_convert_to_set(original_list)
print(unique_set)
"""
addcode(markdown_content, python_code, notebook_file_name='Practical_02_Q8.ipynb')
markdown_content = """

### 5. Mutability and Immutability

#### a. Modify List by Squaring Elements
"""

python_code = """
def square_list_elements(input_list):
    for i in range(len(input_list)):
        input_list[i] **= 2

# Example usage:
numbers_list = [1, 2, 3, 4, 5]
square_list_elements(numbers_list)
print(numbers_list)
"""
addcode(markdown_content, python_code, notebook_file_name='Practical_02_Q9a.ipynb')
markdown_content = """

#### b. Attempt to Modify Tuple Element (Handling Error)
"""
python_code = """
def try_modify_tuple_element(input_tuple, index, new_value):
    try:
        # Tuples are immutable, so modification is not possible
        modified_tuple = input_tuple[:index] + (new_value,) + input_tuple[index + 1:]
        print("Modification successful:", modified_tuple)
    except TypeError:
        print("Error: Tuples are immutable and cannot be modified.")

# Example usage:
my_tuple = (10, 20, 30, 40, 50)
try_modify_tuple_element(my_tuple, 2, 35)
"""
addcode(markdown_content, python_code, notebook_file_name='Practical_02_Q9b.ipynb')
markdown_content = """

### 6. Type Casting

#### a. Convert String Number to Integer and Square
"""
python_code = """
def square_integer_from_string(input_string):
    try:
        number = int(input_string)
        squared_result = number ** 2
        print(f"The squared value of {number} is: {squared_result}")
    except ValueError:
        print("Error: Input is not a valid integer.")

# Example usage:
user_input = input("Enter a number as a string: ")
square_integer_from_string(user_input)
"""
addcode(markdown_content, python_code, notebook_file_name='Practical_02_Q10.ipynb')
markdown_content = """

#### b. Sum Numbers as Strings and Return Integer
"""
python_code = """
def sum_numbers_as_strings(input_list):
    try:
        sum_result = sum(map(int, input_list))
        return sum_result
    except ValueError:
        print("Error: One or more elements in the list is not a valid integer.")

# Example usage:
numbers_as_strings = ['10', '20', '30', '40']
result = sum_numbers_as_strings(numbers_as_strings)
print(f"The sum of the numbers is: {result}")
"""
addcode(markdown_content, python_code, notebook_file_name='Practical_02_Q11.ipynb')
```
